```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from PIL import Image
gen_seq = Image.open("generated_gen_seq_rgb.png")
gen_seq
```

```python
gen_seq = gen_seq.resize((200, 200))
```

```python
gen_seq
```

```python
gen_exp = Image.open("gen_exp.png")
```

```python
gen_exp
```

```python
gen_exp = gen_exp.resize((200, 200))
```

```python
gen_exp
```

```python
pdb = Image.open("pdb_id_dist.png")
pdb = pdb.resize((200, 200))
pdb
```

```python
ims = []
ims.append(pdb)
ims.append(gen_exp)
ims.append(gen_seq)
```

```python
ims = np.array([np.array(im) for im in ims])
```

```python
ims
```

```python
imave = np.average(ims,axis=0)
```

```python
imave
```

```python
result = Image.fromarray(imave.astype('uint8'))
```

```python
result
```

```python
general = np.array(result)
```

```python
general
```

```python
general.shape
```

```python
resh = general.reshape(200, -1)
resh
```

```python
resh.shape
```

```python
gen_df = pd.DataFrame(resh)
gen_df.head()
```

```python
gen_df.shape
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=3)
Kmean.fit(gen_df)
```

```python
Kmean.cluster_centers_
```

```python
Kmean.labels_
```

```python
Kmean.labels_.size
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(gen_df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2'])
```

```python
principalDf['category'] = Kmean.labels_
```

```python
principalDf.head()
```

```python
import seaborn as sns
sns.scatterplot(data = principalDf, x = 'pc1', y = 'pc2', hue = 'category')
```
