```python
import pandas as pd
import numpy as np
```

# Gene Expression Clustering

```python
gen_exp = pd.read_csv('gene_exp_small.csv')
gen_exp = gen_exp.iloc[:, 2:-1]
gen_exp.head()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)
Kmean.fit(gen_exp)
```

```python
labels = Kmean.labels_
```

```python
from sklearn.metrics import davies_bouldin_score, silhouette_score
print(davies_bouldin_score(gen_exp, labels)) 
print(silhouette_score(gen_exp, labels))
```

# PDB ID Clustering

```python
pdb = pd.read_csv('pdb_ids_enc.csv')
pdb = pdb.iloc[:, 1:]
pdb.head()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(pdb)
```

```python
labels = Kmean.labels_
```

```python
print(davies_bouldin_score(pdb, labels)) 
print(silhouette_score(pdb, labels))
```
