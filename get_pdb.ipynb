```python
import pypdb
import pandas as pd
```

```python
gen_seq = pd.read_csv('prostate/gene_sequence.csv')
gen_seq.shape
```

```python
gen_exp = pd.read_csv('prostate/gene_expression.csv')
gen_exp.shape
```

For the sake of computational cost, we take smaller subsets from both the datasets.

```python
gen_exp = gen_exp.iloc[:200, :]
gen_exp.shape
```

The set of genes in the gene sequence is a subset of gene expression

```python
gen_seq = gen_seq.loc[gen_seq['gene'].isin(gen_exp['1'])]
```

```python
gen_seq.head()
```

```python
gen_seq.shape
```

```python
combined_pdb_ids = []
for gene in gen_exp['1']:
    pdb_ids = pypdb.Query(gene).search()
    combined_pdb_ids.append(pdb_ids)
```

```python
combined_pdb_ids
```

```python
pdb_ids = pd.DataFrame(combined_pdb_ids)
pdb_ids.head()
```

```python
pdb_ids.index = gen_exp['1']
```

```python
pdb_ids.head()
```

```python
pdb_ids.to_csv('prostate/pdb_ids.csv')
```

```python
gen_seq.to_csv('prostate/gene_sequence_small.csv')
gen_exp.to_csv('prostate/gene_exp_small.csv')
```
